```python
import pandas as pd
import numpy as np
from plotnine import *
```

```python
df = pd.read_excel("access_count_data_9-14to9-28.xlsx")
```

```python
df.head()
```

```python
df["building_name"] = df["building_name"].astype(str)
```

```python
df["time"] = pd.to_datetime(df["EventDateTime"])
```

```python
df.groupby(pd.Grouper(key="time" , freq="1d"))["access_count"].sum().reset_index()
```

```python
(
    ggplot(df.groupby(pd.Grouper(key="time" , freq="1d"))["access_count"].sum().reset_index())
    + geom_line(aes(x="time" , y="access_count"))
    + theme(axis_text_x=element_text(rotation=45))
    + ggtitle("Access Card Trend")
)
```

```python
df["building_name"] = pd.Categorical(df["building_name"] , categories=df.groupby("building_name")["access_count"].max().sort_values(ascending=False).index)
```

```python
(
    ggplot(df , aes(x="EventDateTime" , y="access_count" , color="building_name"))
    + geom_line()
    + facet_wrap("~ building_name")
    + scale_color_discrete(guide=False)
    + theme(figure_size=(12,12))
)
```
